# Python Tutorial: Converting date strings to datetime format

## Learn how to convert date strings to the `datetime` type in Python and learn why it has become standard practice for working data scientists today

When analyzing datasets with dates and times, there is often a lot to think about. Months of different lengths, different distributions of weekdays and weekends, leap years, and the dreaded timezones are just a few things you may have to consider depending on your context. For this reason, Python has a data type specifically designed for dates and times called `datetime`.

However, in most datasets dates are first represented as strings. So, in this tutorial, you'll learn how to convert date strings to the `datetime` format and see how you can use its powerful set of functionalities to work effectively with complicated chronological data.

### Dealing with different representations of dates

The biggest challenge is often specifying how the date string is specified. For example, `'Wednesday, May 30, 2018'` can also be represented as `'5/30/18'` or as `'05-30-2018'`. These all mean the same thing, but you can probably imagine that the code to convert each of these would be slightly different. Take a moment to examine the code below:

```python
from datetime import datetime

# Define dates as strings
date_str1 = 'Wednesday, May 30, 2018'
date_str2 = '5/30/18'
date_str3 = '05-30-2018'

# Define dates as datetime objects
date_dt1 = datetime.strptime(date_str1, '%A, %B %d, %Y')
date_dt2 = datetime.strptime(date_str2, '%m/%d/%y')
date_dt3 = datetime.strptime(date_str3, '%m-%d-%Y')

# Print converted dates
print(date_dt1)
print(date_dt2)
print(date_dt3)
```

First, you have to import the `datetime` type from the `datetime` module. Then you pass the date string to the `.strptime()` method, followed by Python's `strptime` directives. You can combine directives, special characters (e.g.  `,`, `/`, or `-` as in the cases above), and spaces to match the date string you are trying to parse. As you can see, the resulting `datetime` objects are identical because all three date strings represent the same date.

You can find the full list of directives in the [Python Documentation](http://strftime.org/), but below is a table most relevant to what you saw above:

| Code | Meaning                               | Example        |
|------|---------------------------------------|----------------|
| %A   | Weekday as locale’s full name.        | Wednesday      |
| %a   | Weekday as locale’s abbreviated name. | Wed            |
| %B   | Month as locale’s full name.          | May            |
| %d   | Day of the month.                     | 30             |
| %m   | Month as a number.                    | 05             |
| %Y   | Four-digit year.                      | 2018           |
| %y   | Two-digit year.                       | 18             |

### Converting the date string column

Now that you know how to convert date strings to the `datetime` format, let's learn how to do so when you have a column of date strings in a dataset.

You'll be working with a DataFrame called `eth` that contains some [historical data](https://coinmetrics.io/data-downloads/) on Ether, a cryptocurrency whose blockchain is generated by the Ethereum platform. Your dataset has the following columns:

- `date`: Date, daily at 00:00 UTC
- `txVolume`: Unadjusted measure of the total value, in US dollars, of outputs on the blockchain
- `txCount`: Number of transactions happening on the public blockchain
- `marketCap`: Unit price in US dollars multiplied by the number of units in circulation
- `price`: Opening price in US dollars at 00:00 UTC
- `generatedCoins`: Number of new coins that have been mined into existence
- `exchangeVolume`: The volume, measured by US dollars, at exchanges like GDAX and Bitfinex

Here are the first few rows of your dataset. Note how the dates are represented so you can use the right directives later:

```python
# pre-ex
import pandas as pd
eth = pd.read_csv("data/eth_fin.csv")

# Print head of eth
print(eth.head())
```

And let's confirm that the `date` column needs to be converted:

```python
# Confirm the date column is a string
print(eth.info())
```

The `date` column is indeed a string, which is denoted as an `object` type in Python. You can convert it to a `datetime` type with the `.to_datetime()` method in `pandas`. The console below contains the call to convert the column. Can you complete it by specifying the format with directives you saw earlier?

```python
# Convert date column to datetime
eth['date'] =  pd.to_datetime(eth['date'], format='%Y-%m-%d')

# Confirm the date column is in datetime format
print(eth.info())
```

Did you complete it using `'%Y-%m-%d'`? Great!

### Components of `datetime` objects

Now that you have a `datetime` object, you can get specific components of the date such as the month, day, or year, which are all available as attributes:

```python
# Print datetime attributes
print(eth['date'][0].month)
print(eth['date'][0].day)
print(eth['date'][0].year)
```

Date attributes are often used to group data by a particular time frame. For example, you can see how many ethers were generated on a yearly basis:

```python
from collections import defaultdict
import matplotlib.pyplot as plt

# Initialize defaultdict of type float
yearly_total_coins = defaultdict(float)

# Loop over the rows of eth
for day in eth.iterrows():
    # Get the date
    dates = day[1][0]
    
    # Get the number of coins generated
    num_coins = day[1][5]
    
    # Add the total number of coins to the current value for the year
    yearly_total_coins[dates.year] += num_coins
    
# Print yearly_total_coins
print(yearly_total_coins)

# Visualize aggregated data
plt.bar(range(len(yearly_total_coins)), list(yearly_total_coins.values()), align='center')
plt.xticks(range(len(yearly_total_coins)), list(yearly_total_coins.keys()))
plt.title('# of ethers generated by year')
plt.show()
```

### Adding and subtracting time

Another common case when working with times is to get a date 30, 60, or 90 days in the past from some date. In Python, the `timedelta` object from the `datetime` module is used to represent differences in `datetime` objects. You can create a `timedelta` by passing any number of keyword arguments such as `days`, `seconds`, `microseconds`, `milliseconds`, `minutes`, `hours`, and `weeks` to it.

Once you have a `timedelta` object, you can add or subtract it from a `datetime` object to get another `datetime` object. Try it in the console below:

```python
# Import timedelta from the datetime module
from datetime import timedelta

# Build a timedelta of 59 days, 4 hours, 42 minutes
glanceback = timedelta(days=59, hours=4, minutes=42)

# Get current date and time
right_now = datetime.now()
print(right_now)

# Print the date and time 59 days, 4 hours, 42 minutes ago
print(right_now - glanceback)
```

### Indexing pandas DataFrames with DatetimeIndex

After you've converted the `date` column to the `datetime` format, it is usually a good idea to index your DataFrame by the date, which creates the `DatetimeIndex`. Like the `datetime` type, the `DatetimeIndex` is a special type of index specially designed to work with dates and times. By using the `.set_index()` method with the `inplace` argument set equal to `True`, you can remove the `date` column from your dataset and include it as the index of your DataFrame:

```python
# Don't run this again

# Set index
eth.set_index('date', inplace=True)
print(eth.head(5))
```

Setting the `DatetimeIndex` in your DataFrame allows a much more powerful and flexible set of tools to use when plotting your time series data. For example, `pandas` automatically creates reasonably spaced date labels for the x-axis:

```python
eth.txCount.plot(title='# of transactions')
plt.show()
```

Pretty neat, right?


### Partial string indexing and slicing

Perhaps the most convenient features are partial string indexing and slicing, which allow easy subsetting of your data. Let's say you wanted to take a closer look at the peak around January 2018. The following code uses partial string indexing and slicing to observe the number of transactions between December 2017 and February 2018.

```python
# Use partial string indexing
peak_eth = eth['2017-12': '2018-2']
peak_eth.txCount.plot(title='# of transactions around peak')
```

It's important to note that partial string indexing is inclusive of the boundaries, which is different from typical Python indexing.
